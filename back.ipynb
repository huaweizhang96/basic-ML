```python
import numpy as np

class NN(object):
    def __init__(self, o ,i):
        self.w1, self.b1, self.w2, self.b2, self.w3, self.w4, self.b3 = 0, 0 ,0.1, 0, 0.1, 0.1, 0
        self.x1, self.x2, self.y1, self.y2 = 0, 0, 0, 0
        self.output = o
        self.input = i
        self.prediction = np.zeros(len(o))
        self.learning_rate = 0.00001
        
    def activation_function(self, x, name):
        if name == 'sigmoid':
            return 1/(1 + np.exp(-x))
        elif name == 'softplus':
            return np.log(1 + np.exp(x))
        
    def activation_function_derivative(self, x, name):
        if name == 'sigmoid':
            return self.activation_function(x, 'sigmoid')*(1-self.activation_function(x, 'sigmoid'))
        elif name == 'softplus':
            return self.activation_function(x, 'sigmoid')
        
#     def activation_function(self, x, name):
#         if name == 'sigmoid':
#             return x
#         elif name == 'softplus':
#             return x
        
#     def activation_function_derivative(self, x, name):
#         if name == 'sigmoid':
#             return 1
#         elif name == 'softplus':
#             return 1
        
        
        
        
    def forward(self):
        self.x1 = self.w1*self.input + self.b1
        self.y1 = self.activation_function(self.x1, 'sigmoid')
        self.x2 = self.w2*self.input + self.b2
        self.y2 = self.activation_function(self.x2, 'softplus')
        self.prediction = self.w3*self.y1 + self.w4*self.y2 + self.b3
        
    
    def backward(self):
        lb3 = -2*np.sum(self.prediction - self.output)
        lw3 = -2*np.dot((self.prediction - self.output), self.y1)
        lw4 = -2*np.dot((self.prediction - self.output), self.y2)
        lb1 = -2*np.sum((self.prediction - self.output)*self.w3*self.activation_function_derivative(self.x1, 'sigmoid'))
        lb2 = -2*np.sum((self.prediction - self.output)*self.w4*self.activation_function_derivative(self.x2, 'softplus'))
        lw1 = -2*np.sum((self.prediction - self.output)*self.w3*self.activation_function_derivative(self.x1, 'sigmoid')*self.input)
        lw2 = -2*np.sum((self.prediction - self.output)*self.w3*self.activation_function_derivative(self.x1, 'softplus')*self.input)
        
        
        
        self.b3 = self.b3 - lb3*self.learning_rate
        self.w3 = self.w3 - lw3*self.learning_rate
        self.w4 = self.w4 - lw4*self.learning_rate
        self.b1 = self.b1 - lb1*self.learning_rate
        self.b2 = self.b2 - lb2*self.learning_rate
        self.w1 = self.w1 - lw1*self.learning_rate
        self.w2 = self.w2 - lw2*self.learning_rate
                       
                
    def train(self, times):
        for t in range(times):
            self.forward()
            loss = np.dot((self.prediction - self.output), (self.prediction - self.output))
            print(t, loss)
            self.backward()

            
        
```

```python
i = np.array([1, 2, 3, 4, 5, 6])/10000
o = np.array([1, 2, 3, 4, 5, 6])/10000
nn = NN(o, i) 
```

```python
nn.train(100000)
```
